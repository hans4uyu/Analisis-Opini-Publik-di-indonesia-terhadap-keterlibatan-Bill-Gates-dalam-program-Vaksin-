```python
import pandas as pd
import numpy as np
import re

# ==== Fungsi bantu ====
def clean_text(text):
    """Membersihkan teks dari karakter non huruf dan ubah ke huruf kecil"""
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text.strip()

def stem_word(word):
    """Stemming sederhana berbasis aturan awalan dan akhiran Bahasa Indonesia"""
    prefixes = ['meng', 'meny', 'men', 'mem', 'me',
                'peng', 'peny', 'pen', 'pem', 'di',
                'ke', 'se', 'ber', 'ter', 'per']
    suffixes = ['kan', 'an', 'i', 'lah', 'nya', 'ku', 'mu']

    # Hapus awalan
    for p in prefixes:
        if word.startswith(p) and len(word) > len(p) + 2:
            word = word[len(p):]
            break

    # Hapus akhiran
    for s in suffixes:
        if word.endswith(s) and len(word) > len(s) + 2:
            word = word[:-len(s)]
            break

    return word

def stem_sentence(sentence):
    """Melakukan stemming pada seluruh kata dalam kalimat"""
    cleaned = clean_text(sentence)
    words = cleaned.split()
    stems = [stem_word(w) for w in words]
    return ' '.join(stems)

# ==== Membaca file CSV ====
# Misal dataset.csv berisi kolom: "kalimat"
# kalimat
# Saya menyukai makanan pedas
# Mereka bekerja di kantor pemerintah

df = pd.read_csv('/content/tweets-data/Hasil_Scarping_BillGates_2025.csv')

# Pastikan kolom bernama 'kalimat' ada
if 'full_text' not in df.columns:
    raise ValueError("Kolom 'full_text' tidak ditemukan di file CSV.")

# ==== Proses stemming ====
df['kata_dasar'] = df['full_text'].astype(str).apply(stem_sentence)

# ==== Simpan hasil ke CSV baru ====
df.to_csv('hasil_stem.csv', index=False)

# ==== Tampilkan hasil ====
print(df)
```
