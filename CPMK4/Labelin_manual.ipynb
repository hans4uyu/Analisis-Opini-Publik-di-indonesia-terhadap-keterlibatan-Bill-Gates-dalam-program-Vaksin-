```python
import pandas as pd

df = pd.read_csv("BCA_data_raw_PTP.csv")
# 2. Tambah kolom 'label' kosong
df["label"] = pd.NA
```

```python
def manual_label(df):
    print("== MANUAL LABELING TWEET ==")
    print("Masukkan label:")
    print("1 = positif, 0 = netral, -1 = negatif")
    print("ketik 'skip' untuk lewati, 'stop' untuk berhenti.\n")

    # loop baris
    for i in range(len(df)):
        # kalau sudah ada label (tidak kosong), lewati
        if pd.notna(df.loc[i, "label"]):
            continue

        text = df.loc[i, "content"]
        print(f"\nTweet {i+1}/{len(df)}:")
        print(text)
        label = input("Label (1/0/-1): ").strip()

        if label.lower() == "stop":
            break
        elif label.lower() == "skip":
            continue
        elif label in ["1", "0", "-1"]:
            df.loc[i, "label"] = int(label)
        else:
            print("Input salah, pakai hanya 1 / 0 / -1 / skip / stop.")
            continue

        # autosave tiap kali input label valid
        df.to_csv("BCA_labeled_progress.csv", index=False)
        print("Tersimpan âœ“")

    print("\nSelesai / dihentikan manual. File tersimpan sebagai 'BCA_labeled_progress.csv'.")
    return df
```

```python
df = manual_label(df)
```
